```python
from IPython.core.display import HTML
from IPython.core.pylabtools import figsize
import json, os, glob, csv, json, sys
s = json.load(open('..\\7  Notebook Config Files\\bmh_matplotlibrc.json') )

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams.update(s)

def css_styling():
    styles = open('..\\7  Notebook Config Files\\custom.css', "r").read()
    return HTML(styles)
css_styling()

import numpy as np
import math as m
from tqdm import tqdm
from scipy import signal
from PyQt4 import QtGui
import pylab as plb
from scipy.optimize import curve_fit
from scipy import asarray as ar,exp

figsize(16,9)
```

```python
def openfolder_dialog():
    from PyQt4 import QtGui
    app = QtGui.QApplication([dir])
    fname = QtGui.QFileDialog.getExistingDirectory(None, "Select Folder...", "..\\..\\2  Data\\4  LIF Data\\",QtGui.QFileDialog.ShowDirsOnly)
    return str(fname) + "\\"

def openfile_dialog(folder,prompt):
    from PyQt4 import QtGui
    app = QtGui.QApplication([dir])
    fname = QtGui.QFileDialog.getOpenFileName(None,prompt,folder, filter="All files (*.txt)")
    return str(fname)
#folder = openfolder_dialog()
def scanread(path):
    reader = csv.reader(open(path,'r'),delimiter='\n')
    result = np.array(list(reader))
    dat = {}
    for i in range(0,len(result)-1):
        tmp = result[i][0]
        if i<6:
            dat[tmp.split(' = ')[0]] = tmp.split(' = ')[1]
        elif i == 7:
            cols = tmp.split(', ')
            for i in range(len(cols)):
                dat[cols[i]] = []
        elif i>7:
            for i in range(len(cols)):
                dat[cols[i]].append(float(tmp.split(', ')[i]))
    return dat
```

```python
scans = {}
paths = glob.glob('L:\\Master Drive\\PISCES BIG\\PISCES Share\\2  Data\\4  LIF Data\\SELECTED LIF DATA\\ALLDATA\\' + "*-2015-*.LIF")
savepath = '..\\..\\3  Results\\4  LIF Results\\'
for path in tqdm(paths):
    scans[path.split('\\')[-1].split('.LIF')[0]] = scanread(path)
```

```python
def gaus(x,a,x0,sigma):
    return a*e^{-(x-x0)**2/(2*sigma**2))

def gaussianfit(x,y,scan):
    x = ar(x)
    y = ar(y)
    n = len(x)                          #the number of data
    mean = sum(x*y)/n                   #note this correction
    sigma = m.sqrt(sum(y*(x-mean)**2)/n)        #note this correction
    popt,pcov = curve_fit(gaus,x,y,p0=[1,mean,sigma])
    return gaus(x,*popt)


def mgaus(x,m1,m2,m3,m4):
    return m4+m1*(.375*e^{M0*(x-m2+1.26*B)**2/m3)+ .268*e^{M0*(x-m2+1.44*B)**2/m3)+(.179*e^{M0*(x-m2+1.63*B)**2/m3))+(.107*e^{M0*(x-m2+1.82*B)**2/m3))+(.0536*e^{M0*(x-m2+2.01*B)**2/m3))+(.0178*e^{M0*(x-m2+2.19*B)**2/m3)))

def multigaussianfit(x,y,B,scan):
    x = ar(x)
    y = ar(y)
    n = len(x)                          #the number of data
    mean = sum(x*y)/n                   #note this correction
    sigma = m.sqrt(sum(y*(x-mean)**2)/n)        #note this correction
    popt,pcov = curve_fit(mgaus,x,y,p0=[5,mean+1,sigma+1,.01+1])  #[2,2,4,.5]
    return [mgaus(x,*popt),popt]
```

```python
rejects = {}
for key in tqdm(scans.keys()):
    freq = scans[key]['LIF_Freq']
    signal = scans[key]['LIF_Signal']
    try:
        scans[key]['Single Gaussian Fit'] = gaussianfit(freq,signal,key)
    except:
        rejects[key] = scans[key]
        scans.pop(key,None)
```

```python
def pltfit(freq, sig, fitresults, key, SingleorMulti,save=False):
    plt.gca().cla()    
    plt.plot(freq, sig,'k.',label='Signal')
    plt.plot(freq,fitresults,'co-',label='fit')
    plt.legend()
    plt.title(SingleorMulti + ' Gaussian Fit  ' + key)
    plt.xlabel('Frequency [Ghz]')
    plt.ylabel('Signal')
    plt.ylim(0,max(sig)+.01)
    
    if save==True:
        folder = SingleorMulti + ' Gaussian Fits\\'
        plt.savefig(savepath + folder + key + '.png')

for key in tqdm(scans.keys()):
    freq = scans[key]['LIF_Freq']
    sig = scans[key]['LIF_Signal']
    fitresults = scans[key]['Single Gaussian Fit']
    pltfit(freq,sig,fitresults,key,'Single')
```

```python
refits = {}
for key in tqdm(scans.keys()):
    freq = scans[key]['LIF_Freq']
    signal = scans[key]['LIF_Signal']
    B = (float(scans[key]['Magnetic field'])*3.015+10)/1000
    [fitresults,ms] = multigaussianfit(freq,signal,B,key)
    scans[key]['Fit Parameters'] = ms
    scans[key]['Multi Gaussian Fit'] = fitresults
    pltfit(freq,signal, fitresults, key, 'Multi',True)
    
    [simpfitresults,sms] = multigaussianfit(freq,scans[key]['Single Gaussian Fit'],B,key)
    scans[key]['Simplified Fit Parameters'] = sms
    scans[key]['Simplified Multi Gaussian Fit'] = fitresults
    pltfit(freq,scans[key]['Single Gaussian Fit'], simpfitresults, key, 'Simplified Multi',True)
'''    
    try:
        [fitresults,ms] = multigaussianfit(freq,signal,B,key)
        scans[key]['Fit Parameters'] = ms
        scans[key]['Multi Gaussian Fit'] = fitresults
         
        if ms[1]<0:
            refits[key] = scans[key]
            scans.pop(key,None)        
        else:
            pltfit(freq,signal, fitresults, key, 'Multi',True)
            
    except:
        refits[key] = scans[key]
        scans.pop(key,None)
        '''
```

```python
Tib = []
Vib = []
Bvt = []
Bs = [401.95,582.85, 793.9,1185.85,1276.3,1396.9,1577.8,1788.85,2180.8]
Bss = ['400','600','800','1200','1300','1400','1600','1800','2200']
Tir = {}
Vir = {}
Rvt = {}

for key in tqdm(scans.keys()):
    if float(scans[key]['Radius']) == 0.0:
        Tib.append(scans[key]['Fit Parameters'][2])
        Vib.append(scans[key]['Fit Parameters'][1])
        Bvt.append(float(scans[key]['Magnetic field'])*3.015+10)
    for i in range(len(Bss)):
        if round(float(scans[key]['Magnetic field'])*3.015+10,2) == float(Bs[i]):
            Tir[Bss[i]] = scans[key]['Fit Parameters'][2]
            Vir[Bss[i]] = scans[key]['Fit Parameters'][1]
            Rvt[Bss[i]] = scans[key]['Radius']
```

```python
symbol = {}
symbol['400'] = '+'
symbol['600'] = 'o'
symbol['800'] = 's'
symbol['1200'] = '3'
symbol['1300'] = '*'
symbol['1400'] = ','
symbol['1600'] = '.'
symbol['1800'] = '1'
symbol['2200'] = '2'

figsize(16,18)

for i in range(len(Bss)):
    plt.subplot(211)
    plt.plot(Rvt[Bss[i]],Tir[Bss[i]],symbol[Bss[i]],label=Bss[i] + 'Gauss')
    plt.title('Ti vs R')
    plt.xlabel('Radius [cm]')
    plt.ylabel('Ion Temperature [eV]')
    plt.xlim(-.2,4.7)
    plt.ylim(.1,.95)
    plt.legend()
    
    plt.subplot(212)
    plt.plot(Rvt[Bss[i]],Vir[Bss[i]],symbol[Bss[i]],label=Bss[i] + 'Gauss')
    plt.title('Vi vs R')
    plt.xlabel('Radius [cm]')
    plt.ylabel('Ion Velocity [GHz]')
    plt.xlim(-.2,4.7)
    plt.ylim(0,7)
    plt.legend()
plt.show()
    
```
